# Desarrollo

```python
'''
La idea es comparar la cadena `target` con un sub-cadena de `secuencia` e ir recorriendo toda la cadena de `secuencia
hasta finalmente encontrar la coincidencia
(G)TGATGCTGCTCTTGCTTTGCTGCTGCTTGACAGATTGAAC
(G)CTGCTTGACAGATTGAAC
...
GTGATG(C)TGCTCTTGCTTTGCTGCTGCTTGACAGATTGAAC
----->(G)CTGCTTGACAGATTGAAC
...
GTGATGCTGCTCTTGCTTTGCT(G)CTGCTTGACAGATTGAAC
--------------------->(G)CTGCTTGACAGATTGAAC -> ¡Eureka!
'''


n = len(target) # Longitud de la cadena `target` que estamos buscando
f = len(secuence) - n # La ultima posicion de inicio de la sub-cadena de `secuencia`

# Loop externo: que itera sobre las sub-cadenas de longitud target.len de `secuencia`
for i in range(f):  
  substring = secuence[i:i+n] # Aca recortamos la subcadena que necesitamos

  # print('(',i,')-------------------')
  # print(substring)
  # print(target)
 
  counter = 0  # Inicializar el contador de coincidencias

  # Loop interno: que itera sobra la cadena target Vs. la sub-cadena de secuencia
  for idx, char in enumerate(target):  

    coincidencia_char = char == substring[idx] # Los actuales caracteres coinciden

    # print(char,'=',substring[idx],'?',coincidencia_char)      

    if not coincidencia_char:  # Si no coinciden
      counter = 0         # Se resetea el contador
      break               # Y se rompe el loop para saltar a la siguiente subcadena

    # Si coinciden aumenta el contador
    counter=counter+1

    # print('\t'+str(counter)+',', end='')
    
  # aqui terminamos de comparar y si la coindidencias son sobre toda la subcadena entonces esa es
  if (counter==n):
    print("posición incial:", i)
    print("posición final:", i+n)
    # print(i, substring,i+n) #
    break
      
      
```
